```python
import pandas as pd

import numpy as np

import seaborn as sns

```

```python
iris = sns.load_dataset('iris')
```

```python
iris_df = pd.DataFrame(iris)
```

```python
iris_df
```

```python
iris_df.drop(['species'], axis = 1, inplace = True)
iris_df
```

```python
iris_df['sepal_sum'] = iris_df['sepal_length'] + iris_df['sepal_width']
iris_df
```

```python
iris_df['petal_sum'] = iris_df['petal_length'] + iris_df['petal_width']
iris_df
```

```python
df_sum_stat = pd.DataFrame(iris_df.describe())
```

```python
df_sum_stat
```

```python
df_sum_stat.drop(['count', '25%', '50%', '75%'], axis = 0, inplace = True)
```

```python
df_sum_stat
```

```python
df_sum_stat = df_sum_stat.T
df_sum_stat
```

```python
df_sum_stat = pd.DataFrame(df_sum_stat, index = ['sepal_length', 'sepal_width', 'sepal_sum', 'petal_length', 'petal_width', 'petal_sum'])
df_sum_stat
```
